```python
import pandas as pd 
import numpy as np 
import os
```

```python
path =r"C:\Users\Pocah\OneDrive\Desktop\bootcamp\python"
```

```python
df_customers = pd.read_csv(os.path.join(path, 'Original data', 'customers (1)', 'customers.csv'))
```

```python
df_customers.head()
```

```python
#rename columns 
df_customers.columns = df_customers.columns.str.lower().str.replace(' ', '_')
```

```python
df_customers.head()
```

```python
#checking missing values
print("Missing values:\n", df_customers.isnull().sum())
```

```python
#checking duplicate values 
print("Duplicate rows:", df_customers.duplicated().sum())
```

```python
#fixing data type
df_customers['user_id'] = df_customers['user_id'].astype('int')
```

```python
df_instacart = pd.read_pickle(os.path.join(path, 'Prepared Data', 'ords_prods_merge.pkl'))
```

```python
df_instacart['user_id'] = df_instacart['user_id'].astype('int')
```

```python
#Merged Customer Data with Instacart Data
df_merged = df_instacart.merge(df_customers, on='user_id', how='left')
```
